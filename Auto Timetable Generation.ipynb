```python
import pandas as pd
import random
```

```python
def load_data():
    data = pd.read_csv("timetable_dataset.csv")
    return data
```

```python
def prepare_data(data):
    classes = data["Class"].unique()
    subjects = data["Subject"].unique()
    teachers = data["Teacher"].unique()
    rooms = data["Room"].unique()
    times = data["Time"].unique()

    return classes, subjects, teachers, rooms, times
```

```python
def generate_timetable(data, classes, subjects, rooms, times):

    timetable = []

    for time in times:
        used_teachers = []

        for class_name in classes:

            subject = random.choice(subjects)

            teacher_row = data[data["Subject"] == subject]

            if teacher_row.empty:
                continue

            teacher = teacher_row["Teacher"].values[0]

            while teacher in used_teachers:
                subject = random.choice(subjects)
                teacher_row = data[data["Subject"] == subject]
                teacher = teacher_row["Teacher"].values[0]

            used_teachers.append(teacher)

            room = random.choice(rooms)

            timetable.append({
                "Time": time,
                "Class": class_name,
                "Subject": subject,
                "Teacher": teacher,
                "Room": room
            })

    return timetable
```

```python
def display_timetable(timetable):

    df = pd.DataFrame(timetable)

    print("\nGenerated Timetable\n")
    print(df)

    df.to_csv("generated_timetable.csv", index=False)

    print("\nTimetable saved as generated_timetable.csv")
```

```python
data = load_data()

classes, subjects, teachers, rooms, times = prepare_data(data)

timetable = generate_timetable(data, classes, subjects, rooms, times)

display_timetable(timetable)
```
